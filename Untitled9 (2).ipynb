```python
text = "Python"

print(f"String: {text}")
print(f"Index positions: 0={text[0]}, 1={text[1]}, 2={text[2]}, 3={text[3]}, 4={text[4]}, 5={text[5]}")
print(f"\nNegative indexing: -1={text[-1]}, -2={text[-2]}, -3={text[-3]}")
print(f"\nSlicing:")
print(f"text[0:3] = {text[0:3]}  (characters from index 0 to 2)")
print(f"text[2:5] = {text[2:5]}  (characters from index 2 to 4)")
print(f"text[:3] = {text[:3]}    (first 3 characters)")
print(f"text[3:] = {text[3:]}    (from index 3 to end)")
print(f"text[::2] = {text[::2]}  (every 2nd character)")
```
