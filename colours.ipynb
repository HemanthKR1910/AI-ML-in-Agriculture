```python
colors = ["red", "blue", "green"]

print(f"Original list: {colors}")

colors.append("yellow")
print(f"After append('yellow'): {colors}")

colors.insert(1, "purple")
print(f"After insert(1, 'purple'): {colors}")

colors.remove("purple")
print(f"After remove('purple'): {colors}")

popped = colors.pop()  # Removes last element
print(f"After pop(): {colors}, Popped element: {popped}")

colors.extend(["pink", "orange"])
print(f"After extend(['pink', 'orange']): {colors}")
```
