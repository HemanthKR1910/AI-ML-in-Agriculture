```python
nitrogen_level = float(input("Enter nitrogen level (mg/kg): "))

if nitrogen_level < 100:
  cost = 2000
  print(f"Apply high nitrogen fertilizer and cost is {cost}")
elif 100 <= nitrogen_level <= 250:
  cost = 1000
  print(f"Apply medium nitrogen fertilizer and cost is {cost}")
else:
  cost = 0
  print(f"No fertilizer is needed, soil is healthy and cost is {cost}")
```

```python
rainfall_mm = 45      # Predicted total rainfall (mm)
temperature_c = 36    # Max temperature (°C)
wind_speed_kmh = 35   # Wind speed (km/h)
activity = "planting" # Planned activity: "planting", "spraying", "irrigation"

print(f"5-day forecast: Rainfall {rainfall_mm}mm, Temp {temperature_c}°C, Wind {wind_speed_kmh}km/h")
print(f"Planned activity: {activity}")

if rainfall_mm > 50:
    decision = "DELAY PLANTING"
    reason = "High rainfall predicted - focus on field preparation/drainage"
elif temperature_c > 38:
    decision = "DELAY ACTIVITIES"
    reason = "Extreme heat - prioritize irrigation and shade management"
elif wind_speed_kmh > 40 and activity == "spraying":
    decision = "CANCEL SPRAY OPERATIONS"
    reason = "High winds - risk of spray drift and poor coverage"
else:
    decision = "PROCEED WITH PLANNED ACTIVITIES"
    reason = "Weather conditions favorable for farm work"

print(f"Decision: {decision}")
print(f"Reason: {reason}")
```

```python
pests_per_plant = 12
damage_percent = 18
temp_c = 28
humidity_percent = 75

# Severity classification
if pests_per_plant < 5 and damage_percent < 10:
    severity = "LOW"
    recommendation = "Routine monitoring sufficient."
    cost = "₹200/hectare"
elif (pests_per_plant >= 5 and pests_per_plant <= 15) or (damage_percent >= 10 and damage_percent <= 25):
    severity = "MEDIUM"
    recommendation = "Monitor daily. Consider targeted pesticide spray if severity increases."
    cost = "₹500/hectare"
else:
    severity = "HIGH"
    recommendation = "Immediate action: Apply broad-spectrum pesticide and scout weekly."
    cost = "₹1500/hectare"

# Output
print(f"Pests per plant: {pests_per_plant}, Damage: {damage_percent}%, Temp: {temp_c}°C, Humidity: {humidity_percent}%")
print(f"Severity Level: {severity}")
print(f"Recommendation: {recommendation}")
print(f"Cost of intervention: {cost}")
```

```python
crop = "Wheat"
days_since_sowing = 165

# Maturity ranges (in days)
maturity_ranges = {
    "Rice": (120, 150),
    "Wheat": (140, 180),
    "Cotton": (160, 200)
}

if crop not in maturity_ranges:
    print("Unsupported crop type.")
else:
    min_days, max_days = maturity_ranges[crop]

    if days_since_sowing < min_days:
        status = "NOT READY"
        recommendation = "Continue monitoring growth."
    elif min_days <= days_since_sowing <= max_days:
        status = "READY FOR HARVEST"
        recommendation = "Begin harvesting within next 2 weeks for optimal yield."
    else:
        status = "OVERDUE"
        recommendation = "Harvest immediately to minimize loss."

    print(f"Crop: {crop}, Days since sowing: {days_since_sowing}")
    print(f"Status: {status}")
    print(f"Maturity range: {min_days}-{max_days} days")
    print(f"Recommendation: {recommendation}")
    if days_since_sowing > max_days:
        print("Expected yield loss if delayed: 2% per day")
```
