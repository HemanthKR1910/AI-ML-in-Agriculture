```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor

# -----------------------------
# STEP 1: Enter Historical Dataset
# -----------------------------
print("=== ENTER HISTORICAL TRAINING DATA ===")
n = int(input("How many historical records do you want to enter? "))

temperature = []
rainfall = []
humidity_list = []
soil_moisture = []
yield_list = []

for i in range(n):
    print(f"\nRecord {i+1}")
    temperature.append(float(input("Temperature (°C): ")))
    rainfall.append(float(input("Rainfall (mm): ")))
    humidity_list.append(float(input("Humidity (%): ")))
    soil_moisture.append(float(input("Soil Moisture (%): ")))
    yield_list.append(float(input("Yield (q/ha): ")))

# Create DataFrame
df = pd.DataFrame({
    'Temperature': temperature,
    'Rainfall': rainfall,
    'Humidity': humidity_list,
    'Soil_Moisture': soil_moisture,
    'Yield_q_per_ha': yield_list
})

# -----------------------------
# STEP 2: Train AI Model
# -----------------------------
X = df[['Temperature', 'Rainfall', 'Humidity', 'Soil_Moisture']]
y = df['Yield_q_per_ha']

model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X, y)

# -----------------------------
# STEP 3: Enter Current Field Data
# -----------------------------
print("\n=== ENTER CURRENT FIELD DATA ===")
temp = float(input("Temperature (°C): "))
rain = float(input("Rainfall (mm): "))
humidity = float(input("Humidity (%): "))
soil = float(input("Soil Moisture (%): "))

# -----------------------------
# STEP 4: Predict Yield
# -----------------------------
input_data = pd.DataFrame([[temp, rain, humidity, soil]],
                          columns=['Temperature', 'Rainfall', 'Humidity', 'Soil_Moisture'])

predicted_yield = model.predict(input_data)[0]

print("\n===== AI AGRICULTURE REPORT =====")
print(f"Predicted Ragi Yield: {predicted_yield:.2f} quintals/hectare")

# -----------------------------
# STEP 5: Disease Alerts
# -----------------------------
print("\n===== DISEASE ALERTS =====")

if humidity >= 90 and 20 <= temp <= 25 and rain >= 700:
    print("⚠ Blast Disease Risk: HIGH")
else:
    print("✓ Blast Disease Risk: LOW")

if humidity >= 85 and 24 <= temp <= 30:
    print("⚠ Leaf Spot Risk: HIGH")
else:
    print("✓ Leaf Spot Risk: LOW")

if soil >= 38 and rain >= 850:
    print("⚠ Root Rot Risk: HIGH")
else:
    print("✓ Root Rot Risk: LOW")

# -----------------------------
# STEP 6: Irrigation Advisory
# -----------------------------
if soil < 25:
    print("💧 Irrigation Needed: YES")
elif soil < 30:
    print("💧 Irrigation Needed: Moderate")
else:
    print("💧 Irrigation Needed: NO")
```
