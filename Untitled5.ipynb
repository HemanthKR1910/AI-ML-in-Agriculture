```python
rainfall_mm = 45      # Predicted total rainfall (mm)
temperature_c = 36    # Max temperature (°C)
wind_speed_kmh = 35   # Wind speed (km/h)
activity = "planting" # Planned activity: "planting", "spraying", "irrigation"

print(f"5-day forecast: Rainfall {rainfall_mm}mm, Temp {temperature_c}°C, Wind {wind_speed_kmh}km/h")
print(f"Planned activity: {activity}")

if rainfall_mm > 50:
    decision = "DELAY PLANTING"
    reason = "High rainfall predicted - focus on field preparation/drainage"
elif temperature_c > 38:
    decision = "DELAY ACTIVITIES"
    reason = "Extreme heat - prioritize irrigation and shade management"
elif wind_speed_kmh > 40 and activity == "spraying":
    decision = "CANCEL SPRAY OPERATIONS"
    reason = "High winds - risk of spray drift and poor coverage"
else:
    decision = "PROCEED WITH PLANNED ACTIVITIES"
    reason = "Weather conditions favorable for farm work"

print(f"Decision: {decision}")
print(f"Reason: {reason}")
```
