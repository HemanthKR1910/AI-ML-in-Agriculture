```python
sentence = "hello world python"

print(f"Original: {sentence}")
print(f"Upper: {sentence.upper()}")
print(f"Lower: {sentence.lower()}")
print(f"Capitalize: {sentence.capitalize()}")
print(f"Title Case: {sentence.title()}")
print(f"Replace 'python' with 'java': {sentence.replace('python', 'java')}")
print(f"Split into list: {sentence.split()}")
print(f"Count 'o': {sentence.count('o')}")
print(f"Starts with 'hello': {sentence.startswith('hello')}")
print(f"Ends with 'python': {sentence.endswith('python')}")
```
