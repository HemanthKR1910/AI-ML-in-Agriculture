```python
nitrogen_level = float(input("Enter nitrogen level (mg/kg): "))

if nitrogen_level < 100:
  cost = 2000
  print(f"Apply high nitrogen fertilizer and cost is {cost}")
elif 100 <= nitrogen_level <= 250:
  cost = 1000
  print(f"Apply medium nitrogen fertilizer and cost is {cost}")
else:
  cost = 0
  print(f"No fertilizer is needed, soil is healthy and cost is {cost}")
```
