```python
sentence="i like agriculture"
vowels="aeiou" # Define vowels as a string of individual characters
total_vowel_count = 0
for v in vowels:
    total_vowel_count += sentence.count(v) # Count each vowel in the sentence
print(f"Total number of vowels: {total_vowel_count}")
```

```python
numbers=[1,3,5,7]
n=15
# Find numbers from 1 to n that are not in the 'numbers' list
missing = [i for i in range(1, n + 1) if i not in numbers]
print(missing)
```
