```python
# Precision Irrigation System
# Inputs: crop type, soil moisture %, temperature

crop = input("Enter crop type (rice / wheat / sugarcane): ").lower()
moisture = float(input("Enter soil moisture percentage: "))
temperature = input("Enter temperature (hot / moderate / cool): ").lower()

if crop == "rice":
    if moisture < 50 and temperature == "hot":
        print("Irrigation needed for rice 🌾")
    elif moisture < 60 and temperature in ["moderate", "cool"]:
        print("Irrigation needed for rice 🌾")
    else:
        print("No irrigation needed for rice")

elif crop == "wheat":
    if moisture < 40 and temperature == "hot":
        print("Irrigation needed for wheat 🌾")
    elif moisture < 30 and temperature in ["moderate", "cool"]:
        print("Irrigation needed for wheat 🌾")
    else:
        print("No irrigation needed for wheat")

elif crop == "sugarcane":
    if moisture < 70:
        print("Irrigation needed for sugarcane 🎋")
    else:
        print("No irrigation needed for sugarcane")

else:
    print("Invalid crop type entered")
```
